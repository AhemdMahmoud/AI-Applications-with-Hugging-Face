```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("translation", model="facebook/nllb-200-distilled-600M")
```

```python
massage="Ahmed mahmoud is the best programer  in the regoin"
```

```python
test_g=pipe(massage,src_lang="en",tgt_lang="ars_Arab")
#ars_Arab ,arb_Arab
```

```python
test_g[0]["translation_text"]
```
