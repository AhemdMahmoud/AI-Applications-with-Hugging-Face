```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text2text-generation", model="facebook/blenderbot-400M-distill")
```

```python
test=pipe("what are you can do ")
test[0]["generated_text"]
```

```python
# build function 
def chat(text):
    text=pipe(text)
    text=text[0]['generated_text']
    return text
```

```python
from transformers import pipeline
# Function to add a message and generate a response
def add_message(history, text):
    history.append({"role": "user", "content": text})  # Add user input to history
    response = pipe(history[-1]["content"])  # Generate a response based on the latest message
    generated_text = response[0]['generated_text']  # Extract generated text
    history.append({"role": "assistant", "content": generated_text})  # Add the response to history
    return history

# Example usage
history = []
history = add_message(history, "hello")  # Start conversation
print(history)
```
